## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Creating an instance of label encoder
label_encoder = LabelEncoder()

# list of columns that can be encoded (2 value types)
# Encode the loan_status as an integer: 1="low_risk", 0="high_risk"
le_cols = ["loan_status", "pymnt_plan", "initial_list_status", "application_type", "hardship_flag", "debt_settlement_flag"]
for col in le_cols:
    label_encoder.fit(df[col])
    # Encode the column as an integer
    df[col] = label_encoder.transform(df[col])

# Convert 'issue_d' and 'next_pymnt_d' to month number
df["issue_d"] = pd.DatetimeIndex(df["issue_d"]).month
df["next_pymnt_d"] = pd.DatetimeIndex(df["next_pymnt_d"]).month

# Perform a binary encoding on the `home_ownership` and `verification_status` columns using the Pandas `get_dummies()` function.
df = pd.get_dummies(df, columns=["home_ownership", "verification_status"])

# Preview the data
df.tail(100)
```

```python
file_path = Path("../Resources/LoanStats_2019Q1_encoded.csv")
df.to_csv(file_path, index=False)
```

## Split the Data into Training and Testing

```python
# Create our features
# YOUR CODE HERE
X = X = df.drop("loan_status", axis=1) 
# Create our target
# YOUR CODE HERE
y = df["loan_status"].to_frame()
```

```python
# Check the balance of our target values
# YOUR CODE HERE
y['loan_status'].value_counts()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
# YOUR CODE HERE
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y['loan_status'], random_state=1)
Counter(y_train)
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# Create the StandardScaler instance
# YOUR CODE HERE
from sklearn.preprocessing import StandardScaler
# YOUR CODE HERE
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
# YOUR CODE HERE
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training and testing data
# YOUR CODE HERE
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
# YOUR CODE HERE
from imblearn.ensemble import BalancedRandomForestClassifier
brf = BalancedRandomForestClassifier(n_estimators=1000, random_state=1)
brf.fit(X_train_scaled, y_train)
```
